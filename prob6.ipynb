```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['species'] = pd.Categorical.from_codes(data.target, data.target_names)

# Calculate the mean of each feature grouped by species
mean_values = df.groupby('species').mean()

# Rename columns to indicate sepal and petal sizes
mean_values.columns = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(mean_values, annot=True, cmap="YlGnBu", cbar=True)
plt.title("Mean Sepal and Petal Sizes by Species")
plt.show()
```
